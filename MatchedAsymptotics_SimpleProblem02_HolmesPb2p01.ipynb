```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
"""
Problem 2.1 from Holmes book. Friedrichs model for boundary layer in viscous fluid.
"""

x = np.logspace(-3,0,100)

eps = 0.4
a = 0.75

exactSol = (1-a)/(1-np.exp(-1/eps))*(1-np.exp(-x/eps)) + a*x

innerSol = (1-a)*(1-np.exp(-x/eps))
outerSol = a*x+(1-a)
compSol = a*x + (1-a)*(1-np.exp(-x/eps))

bettercompSol = a*x + (1-a)*(1-np.exp(-x/eps)) + (1-a)*np.exp(-1/eps) 
"""
page 87 of Holmes' book discusses how to add transcendentally small terms to the outer
solution when the outer solution seems to contain only one term, the y0 term. All other
terms like y1, y2, etc vanish. Then, we need transcendentally small terms because the 
y0 term does not satisfy the BC at x=0, i.e., what seems like an exact solution cannot 
be the exact solution.

This can be seen clear from the plots for larger values of eps, like eps=0.4. 
Between the exact solution and the composite solution, there is a small error as 
x approaches 1. 

Actually, from the plot, it looks like the addition of the term is pointless. It improves the
error at x=1 but makes it worse at x=0. What do you expect by adding a constant term anyway!
Not clear why Holmes added this extra term. That may be because I am wrong about the form of the
improved composite Solution. I used Eq. 2.56 of Holmes as the basis of this equation

bettercompSol = a*x + (1-a)*(1-np.exp(-x/eps)) + (1-a)*np.exp(-1/eps)

However, 2.56 is only the improved outer solution. Need to obtain the appropriate composite
solution and modify the expression for bettercompSol.
"""
f, (ax1) = plt.subplots(1, figsize=(9,6))
ax1.plot(x, outerSol, 'b--', label='outer')
ax1.plot(x, innerSol, 'y--', label='inner')
ax1.plot(x, compSol, 'r-', label='composite')
ax1.plot(x, bettercompSol, 'k-', label='better comp')
ax1.plot(x, exactSol, 'g-', label='exact')
ax1.set_xlabel('$x$',fontsize=16)
ax1.set_ylabel('$y_0(x)$', fontsize=16)
ax1.set_xscale('log')
ax1.legend(loc='lower right')

f, (ax1) = plt.subplots(1, figsize=(9,6))
ax1.plot(x, outerSol, 'b--', label='outer')
ax1.plot(x, innerSol, 'y--', label='inner')
ax1.plot(x, compSol, 'r-', label='composite')
ax1.plot(x, bettercompSol, 'k-', label='better comp')
ax1.plot(x, exactSol, 'g-', label='exact')
ax1.set_xlabel('$x$',fontsize=16)
ax1.set_ylabel('$y_0(x)$', fontsize=16)
ax1.set_xscale('linear')
ax1.legend(loc='lower right')
#f.show()

f, (ax1) = plt.subplots(1, figsize=(9,6))
ax1.plot(x, exactSol-compSol, 'r--', label='comp, error')
ax1.plot(x, exactSol-bettercompSol, 'k--', label='better comp, error')
ax1.set_xlabel('$x$',fontsize=16)
ax1.set_ylabel('$y_{exact}-y_0(x)$', fontsize=16)
ax1.set_xscale('linear')
ax1.legend(loc='lower right')
```
