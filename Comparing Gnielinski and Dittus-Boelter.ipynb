```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

ReD = np.logspace(3.37,6,300)
f  = (0.79*np.log(ReD)-1.64)**(-2)
Pr = 1.75
NuD_G = (f/8)*(ReD-1000)*Pr/(1+12.7*np.sqrt(f/8)*(Pr**(2/3)-1))

NuD_DB = 0.023*ReD**0.8*Pr**0.4

f, (ax1) = plt.subplots(1, figsize=(9,6))
ax1.plot(ReD, NuD_G, 'r--', label='Gnielinski')
ax1.plot(ReD, NuD_DB, 'b--', label='Dittus-Boelter')
ax1.set_xscale('log')
ax1.set_yscale('log')

ax1.set_xlabel(r'$Re_D$', fontsize=16)
ax1.set_ylabel(r'$Nu_D$', fontsize=16)
ax1.set_xticklabels(ax1.get_xticks(), fontsize=16)
ax1.set_yticklabels(ax1.get_yticks(), fontsize=16)

ax1.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))

ax1.legend(loc='lower right', fontsize=14)

plt.show()
```
