```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

x = np.logspace(-3,0,200)
eps = 0.1
a = 0.75

exactSol = a*x+(1-a)*(1-np.exp(-x/eps))/(1-np.exp(-1/eps))

innerSol_0 = (1-a)*(1-np.exp(-x/eps))
outerSol_0 = a*x+1-a
compSol_0 = a*x+(1-a)*(1-np.exp(-x/eps))

innerSol_1 = a*x
outerSol_1 = 0.0
compSol_01 = a*x+(1-a)*(1-np.exp(-x/eps))

f, (ax1) = plt.subplots(1, figsize = (9,6))
ax1.plot(x, innerSol_0, 'g--', label='inner')
ax1.plot(x, outerSol_0, 'y--', label='outer')
ax1.plot(x, compSol_0, 'r-', label='composite')
ax1.plot(x, exactSol, 'b--', label='exact')

ax1.set_xlabel(r'$x$', fontsize=16)
ax1.set_ylabel(r'$y_0(x), Y_0(x), y_c(x), y_{exact}(x)$', fontsize=16)
ax1.set_xticklabels(ax1.get_xticks(), fontsize=16)
ax1.set_yticklabels(ax1.get_yticks(), fontsize=16)

ax1.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))

ax1.set_xscale('log')
ax1.legend(loc='upper left', fontsize=14)

plt.title('Holmes, Pb. 2.1, Friedrichs boundary layer')

plt.show()
```
