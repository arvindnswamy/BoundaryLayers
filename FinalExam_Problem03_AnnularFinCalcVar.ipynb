```
import numpy as np
from scipy.special import kn, iv
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['lines.linewidth'] = 2
matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.size'] = 36


B = 2.0 # ratio b/a or r2/r1 - fin outer radius/fin root radius
beta = 2.0 # a*np.sqrt(2h/(k*delta))

mL = beta*(B-1.0)

rho = np.linspace(1.0, B, 100)
 
thetaAct = (iv(0,beta*rho)/iv(1,beta*B)+kn(0,beta*rho)/kn(1,beta*B))/(iv(0,beta)/iv(1,beta*B)+kn(0,beta)/kn(1,beta*B))

a0 = 0.5*(5*B**4/12 - B**3 + B**2/2 + B/3 - 1/4)/((B-1)**5*(11*B+5)/120 + (1/beta**2)*(B**4/12-B**2/2+2*B/3-1/4))

thetaApprox = 1 - 0.5*a0*(1-rho**2 -2*B*(1-rho))

f, (ax1) = plt.subplots(1, figsize=(12,6))
ax1.plot(rho, thetaAct, label='actual')
ax1.plot(rho, thetaApprox, label='approximate')
ax1.set_xlabel(r'$ \rho = r/a$')
ax1.set_ylabel(r'$\theta_{act}, \theta_{approx}$')
ax1.legend(loc='upper right')
plt.show()
```
