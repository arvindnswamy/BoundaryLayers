```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt

zloc = np.array([25.0, 50.0, 100.0, 200.0])
z = np.linspace(1,210.0, 200)

vonkarmank = 0.4 
vonkarmanz0 = 0.08 #in cm

monteithVel = np.array([[178.0, 206.0,224.0,242.0], [256.0, 288, 316, 368], [293.0, 331., 362., 400.],[204, 223,249., 263.]])

print(np.shape(zloc))
print(np.shape(monteithVel))

f,(ax1) = plt.subplots(1, figsize=(9,6))
ax1.plot(zloc, monteithVel[0,:], 'o')
ustar = 12.7 #in cm/s
ax1.plot(z, (ustar/vonkarmank)*np.log(z/vonkarmanz0))

ax1.plot(zloc, monteithVel[3,:], 'o')
ustar = 13.7 #in cm/s
ax1.plot(z, (ustar/vonkarmank)*np.log(z/vonkarmanz0))

ax1.plot(zloc, monteithVel[1,:], 'o')
ustar = 17.7 #in cm/s
ax1.plot(z, (ustar/vonkarmank)*np.log(z/vonkarmanz0))

ax1.plot(zloc, monteithVel[2,:], 'o')
ustar = 20.5 #in cm/s
ax1.plot(z, (ustar/vonkarmank)*np.log(z/vonkarmanz0))

ax1.set_xticklabels(ax1.get_xticks(), fontsize=16)
ax1.set_yticklabels(ax1.get_yticks(), fontsize=16)

ax1.set_xlabel('vertical position (cm)', fontsize=16)
ax1.set_ylabel('Wind velocity (cm/s)', fontsize=16)
plt.title('Table 2, Monteith, 1957', fontsize=16)

plt.show()
```
